```python
import json
import flatten_json
import pandas as pd
```

```python
f = open("data/logfile_2024_04_12-05.ljson", "r")

data = []
line_arr = f.readlines()
for row in line_arr:
    try:
        j_row = json.loads(row.strip())
        if "type" in j_row and j_row["type"] == "meas":
            j_flat = flatten_json.flatten(j_row)
            time_str = float(j_row["time"]) * 1e6
            data.append(j_flat)
    except BaseException as e:
        print(e)
        
```

```python
df = pd.DataFrame(data)
df.set_index(df["time"])
del df["type"]
```
